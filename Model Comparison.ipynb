```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
SVM_HIST = np.load("Results/SVM_HIST.npy")
SVM_HOG = np.load("Results/SVM_HOG.npy")

NN_HIST = np.load("Results/NN_HIST.npy")
NN_HOG = np.load("Results/NN_HOG.npy")

CNN = np.load("Results/CNN.npy")
```

```python
plt.figure(figsize=(20,10))

plt.plot(SVM_HIST[:,0], SVM_HIST[:,1])
a1 = round(metrics.auc(SVM_HIST[:,0], SVM_HIST[:,1]),4)
print(a1)

plt.plot(SVM_HOG[:,0], SVM_HOG[:,1])
a2 = round(metrics.auc(SVM_HOG[:,0], SVM_HOG[:,1]), 4)
print(a2)

plt.plot(NN_HIST[:,0], NN_HIST[:,1])
a3 = round(metrics.auc(NN_HIST[:,0], NN_HIST[:,1]), 4)
print(a3)

plt.plot(NN_HOG[:,0], NN_HOG[:,1])
a4 = round(metrics.auc(NN_HOG[:,0], NN_HOG[:,1]), 4)
print(a4)

plt.plot(CNN[:,0], CNN[:,1])
a5 = round(metrics.auc(CNN[:,0], CNN[:,1]), 4)
print(a5)

plt.plot([0,1],[0,1], linestyle='dashed')

plt.legend(["SVM Histogram: " + str(a1),
            "SVM HOG: " + str(a2),
            "NN Histogram: " + str(a3),
            "NN HOG: " + str(a4),
            "CNN: " + str(a5)], fontsize=20)

plt.title("ROC Curve Comparison For All Models", fontsize=20)
```
